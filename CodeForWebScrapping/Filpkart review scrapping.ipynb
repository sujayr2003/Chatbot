```python
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import time
from tqdm import tqdm

driver=webdriver.Chrome()
driver.get("https://www.flipkart.com/search?q=mobiles")
html_data_1=BeautifulSoup(driver.page_source,'html.parser')
#html_data_1.find('span',{'class':'s-pagination-item s-pagination-disabled'}).text
No_of_pages=10
titles=[]
images=[]
ratings=[]
prices=[]
product_urls=[]
for i in tqdm(range(int(No_of_pages))):
   url = 'https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(i+1)
   driver.get(url)
   time.sleep(3)
   html_data_2=BeautifulSoup(driver.page_source,'html.parser')
   products=html_data_2.find_all('div',{'class':'cPHDOP col-12-12'})
   for product in products:
      title=product.find('div',{'class':'KzDlHZ'})
      if title == None:
         title='No Title'
      else:
        title=product.find('div',{'class':'KzDlHZ'}).text
      titles.append(title)
      if product.find('img')==None:
         img='No Image'

      else:
         img = product.find('img')['src']
      images.append(img)
      rating=product.find('div',{'class':'XQDdHH'})
      if rating == None:
         rating='No Rating'
      else:
         rating=product.find('div',{'class':'XQDdHH'}).text
      ratings.append(rating)
      price = product.find('div',{'class':'Nx9bqj _4b5DiR'})
      if price==None:
         price='No Price'
      else:
         price = product.find('div',{'class':'Nx9bqj _4b5DiR'}).text
      prices.append(price)
      product_url=product.find('a',{'class':'CGtC98'})
      if product_url==None:
         product_url='No URL'
      else:
         product_fullurl="https://flipkart.com"+product_url['href']
      product_urls.append(product_fullurl)
```

```python
data = pd.DataFrame({"titles":titles,"images":images,"ratings":ratings,"prices":prices,"product_urls":product_urls})
data.to_csv(r"C:\Users\sujay\Desktop\Chatbot\CSVFilesObtainedAfterWebscrapping\Flipkart\Products.csv")
```
